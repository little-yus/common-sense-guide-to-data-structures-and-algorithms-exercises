```python
# 2. Optimize function using memoization
def golomb_original(n):
    if n == 1:
        return 1
    return 1 + golomb_original(n - golomb_original(golomb_original(n - 1)))
```

```python
[golomb_original(n) for n in range(1, 40)]
```

```python
def golomb(n, memo = {}):
    if n == 1:
        return 1
    
    if n in memo:
        return memo[n]
    
    memo[n] = 1 + golomb(n - golomb(golomb(n - 1)))
    return memo[n]
```

```python
[golomb(n) for n in range(1, 100)]
```

```python
# 3. Here is a solution to the “Unique Paths” problem from an exercise in the previous chapter. Use memoization to improve its efficiency.
def unique_paths_original(rows, columns):
    if rows == 1 or columns == 1:
        return 1
    
    return unique_paths_original(rows - 1, columns) + unique_paths_original(rows, columns - 1)
```

```python
for rows in range(1, 10):
    for cols in range(1, 10):
        print(f'unique_paths_original({rows}, {cols}) = {unique_paths_original(rows, cols)}')
```

```python
def unique_paths(rows, columns, memo = {}):
    if rows == 1 or columns == 1:
        return 1
    
    if (rows, columns) in memo:
        return memo[(rows, columns)]
    
    # we utilize property of the original function that unique_paths(rows, columns) == unique_paths(columns, rows)
    if (columns, rows) in memo:
        return memo[(columns, rows)]
    
    memo[(rows, columns)] = unique_paths(rows - 1, columns) + unique_paths(rows, columns - 1)
    
    return memo[(rows, columns)]
```

```python
for rows in range(1, 10):
    for cols in range(1, 10):
        print(f'unique_paths({rows}, {cols}) = {unique_paths(rows, cols)}')
```
